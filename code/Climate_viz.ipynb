```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('ticks')
sns.set_context('poster')
```

```python
# get data from Marcott et al. (10.1126/science.1228026) 
df = pd.read_csv('/Users/urai/Downloads/Marcott.csv')
df.head(n=5)

# get data from https://commons.wikimedia.org/wiki/File:All_palaeotemps.svg#cite_note-11
df_n = pd.read_csv('/Users/urai/Downloads/NGRIP.csv')
df_n.head(n=5)
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,8))
sns.lineplot(data=df, x='year', y='temp', color='blue', ax=ax)
sns.scatterplot(data=df_n, x='year', y='temp', marker='.', color='darkblue', ax=ax)
ax.set(xlim=[-100000, 2022], ylim=[-8, 5], 
       xlabel='', ylabel='Temparature anomaly ($\Delta$$^\circ$C)',
      xticks=[-100000, -75000, -50000, -25000, 0],
      yticks=[-6, -3, 0, 3])
ax.axhline(y=0, color='.15', linestyle='--', zorder=-100)
sns.despine(trim=True)
fig.savefig('temphist_1.pdf')
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,8))
sns.lineplot(data=df, x='year', y='temp', color='blue', ax=ax)
sns.scatterplot(data=df_n, x='year', y='temp', marker='.', color='darkblue', ax=ax)
ax.set(xlim=[-10000, 2100], ylim=[-3, 5], 
       xlabel='', ylabel='Temparature anomaly ($\Delta$$^\circ$C)',
      xticks=[-8000, -6000, -4000, -2000, 0, 2000])
ax.axhline(y=0, color='.15', linestyle='--', zorder=-100)
sns.despine(trim=True)
fig.savefig('temphist_2.pdf')
```

```python
# data from Our World in Data: https://ourworldindata.org/explorers/ipcc-scenarios?facet=none&Metric=Temperature+increase&Rate=Per+capita&Region=Global&country=SSP1+-+Baseline~SSP2+-+Baseline~SSP3+-+Baseline~SSP4+-+Baseline~SSP5+-+Baseline
df2 = pd.read_csv('/Users/urai/Downloads/ipcc.csv', columns=['CO2 emissions', 'Temperature increase', 'Scenario'])
df2 = df2[df2.Scenario.str.contains('Baseline')]
df2.head(n=5)
df2.Scenario.unique()
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,8))
sns.lineplot(data=df2, x='Year', y='Temperature', hue='Scenario', 
             hue_order=['SSP1 - Baseline', 'SSP2 - Baseline', 
       'SSP4 - Baseline', 'SSP5 - Baseline'],
             palette='autumn_r', ax=ax, legend=False)
ax.set(xlim=[1990, 2100], ylim=[-3, 5], 
       xlabel='', ylabel='Temparature anomaly ($\Delta$$^\circ$C)',
      xticks=[2000, 2050, 2100])
ax.axhline(y=0, color='.15', linestyle='--', zorder=-100)
sns.despine(trim=True)
fig.savefig('temphist_3.pdf')
```

```python
fig, ax = plt.subplots(1,1, figsize=(14,8))

sns.scatterplot(data=df_n, x='year', y='temp', marker='.', color='darkblue', ax=ax)

sns.lineplot(data=df2, x='Year', y='Temperature', hue='Scenario', 
             hue_order=['SSP1 - Baseline', 'SSP2 - Baseline', 
       'SSP4 - Baseline', 'SSP5 - Baseline'],
             palette='autumn_r', ax=ax, legend=False)

ax.set(xlim=[-10000, 2200], ylim=[-8, 5], 
       xlabel='', ylabel='Temparature anomaly ($\Delta$$^\circ$C)',
      xticks=[-100000, -75000, -50000, -25000, 0],
      yticks=[-6, -3, 0, 3, 6])
ax.axhline(y=0, color='.15', linestyle='--', zorder=-100)
sns.despine(trim=True)
fig.savefig('temphist_3.pdf')
```
